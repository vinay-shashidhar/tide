<h2>Feature Selection</h2>

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

y = df[['output']]
X = df[cols]
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X, y)
sorted_idx = rf.feature_importances_.argsort()
plt.barh(cols[sorted_idx], rf.feature_importances_[sorted_idx])
print('Explained variance :',np.sum(rf.feature_importances_[sorted_idx][3:8]))
print('Selected features' ,cols[sorted_idx][3:8])
selected_cols = cols[sorted_idx][3:8]
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GroupShuffleSplit


y = df[['output']]
X = df

rus = SMOTE(random_state=42)

# X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)


train_inds, test_inds = next(GroupShuffleSplit(test_size=.20, n_splits=2, random_state = 7).split(df, groups=df['receipt_id']))

trainSet = df.iloc[train_inds]
testSet = df.iloc[test_inds]

# trainSet = X_train.copy()
# trainSet.loc[:,'output'] = y_train
# testSet = X_test.copy()
# testSet.loc[:,'output'] = y_test

print('Unique users in train',len(trainSet.receipt_id.unique()))
print('Unique users in test',len(testSet.receipt_id.unique()))

X_train = trainSet[selected_cols]
X_test = testSet[selected_cols]
y_train = trainSet[['output']]
y_test = testSet[['output']]

X_train = np.asarray(X_train)
X_test = np.asarray(X_test)

print('Positive same in train set',np.mean(y_train))
print('Positive same in test set',np.mean(y_test))
```

```python
histPLot = pd.DataFrame(trainSet.groupby(by="receipt_id").output.count()).reset_index()
histPLot.hist(bins=100, weights=np.ones_like(histPLot[histPLot.columns[0]]) * 100. / len(histPLot))
print(histPLot.describe())
print('Users with more than 3 transaction : ',np.sum(histPLot.output>3))
```

<h2>Modelling</h2>

```python
space={'max_depth': hp.quniform("max_depth", 5, 10, 1),
    'reg_alpha' : hp.quniform('reg_alpha', 40,80,1),
    'reg_lambda' : hp.uniform('reg_lambda', 0,1),
    'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
    'min_child_weight' : hp.quniform('min_child_weight', 0, 5, 1),
    'n_estimators': hp.quniform("n_estimators", 50, 300, 20),
    'subsample': hp.uniform('subsample', 0.95, 1),
    'seed':hp.choice('seed','0'),
    'subsample_freq':hp.choice('subsample_freq','0'),
    'num_leaves': hp.quniform('num_leaves', 31, 40, 1)
}

def objective(space):
    clf=lgb.LGBMClassifier(
                    n_estimators =int(space['n_estimators']), 
                    max_depth = int(space['max_depth']),
                    reg_alpha = int(space['reg_alpha']),
                    min_child_weight=int(space['min_child_weight']),
                    colsample_bytree=(space['colsample_bytree']),
                    reg_lambda=(space['reg_lambda']),
                    subsample = (space['subsample']),
                    objective ="binary",
                    num_leaves = int((space['num_leaves'])),
                    verbose=-1)

    evaluation = [( X_train, y_train), ( X_test, y_test)]

    clf.fit(X_train, y_train,
            eval_set=evaluation, eval_metric="auc",
            early_stopping_rounds=10)


    pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, pred>0.5)
#     print ("SCORE:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }

trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 10,
                        trials = trials,
                        verbose=-1)

print("The best hyperparameters are : ","\n")
print(best_hyperparams)


```

```python
best_hyperparams['max_depth']=int(best_hyperparams['max_depth'])
best_hyperparams['n_estimators']=int(best_hyperparams['n_estimators'])
best_hyperparams['num_leaves']=int(best_hyperparams['num_leaves'])
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt

model = lgb.LGBMClassifier(**best_hyperparams)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import PrecisionRecallDisplay

print('-----Training Result ---')
# y_pred = model.predict(X_train)
y_pred = model.predict_proba(X_train)[:,1]
trainSet.loc[:,'predictedLGB'] = y_pred
# bestScore,T = bestThresshold(y_train,y_pred)
T = 0.3
y_pred = y_pred>T
print(classification_report(y_train, y_pred))
print(roc_auc_score(y_train, y_pred))
print(confusion_matrix(y_train,y_pred))
print('MCC : ',matthews_corrcoef(y_train, y_pred)) #MCC
# plot_confusion_matrix(model,X_train,y_train)
print('-----Test Result ---')
# y_pred = model.predict(X_test)
y_pred = model.predict_proba(X_test)[:,1]
testSet.loc[:,'predictedLGB'] = y_pred
y_pred = y_pred>T
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
print('MCC : ',matthews_corrcoef(y_test, y_pred)) #MCC
print('ROC AUC Score', roc_auc_score(y_test, y_pred))
plot_confusion_matrix(model,X_test,y_test)
```
